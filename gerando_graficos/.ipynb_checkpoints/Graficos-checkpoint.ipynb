```python
import pandas as pd
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
import re
```

## Funcoes

```python
## Captura os valores da media de cada seed
def media_erro_df(df):
    vetor_media=[]
    somas_media=0
    cont=0
    
    for i in range(len(df)): 
        if(df.loc[i,"Likelihood position X (m)"] == 'MÃ©dia'):
            vetor_media.append(df.loc[i,"Erro"])
            somas_media=somas_media+df.loc[i,"Erro"]
            cont=cont+1
    media=somas_media/cont
    return media, vetor_media

    
```

```python
def criando_df(dataframe_bruto, nome_movimento):
    nos=['3','6','9']
    velocidade=['1ms','15ms','30ms']
    
    lista_de_lista=[]
    for coluna_nos in nos:
        for coluna_velocidade in velocidade:
            ## Nome da coluna do dataframe bruto
            nome_coluna='Wireless'+coluna_nos+'h'+'_'+coluna_velocidade+'_'+nome_movimento
            lista_erro=dataframe_bruto[nome_coluna]
            
            ## Renomeia a velocidade de 1ms para 1 m/s
            coluna_velocidade = re.sub(r'ms',' m/s',coluna_velocidade)
            
            for erro in lista_erro:
                linha=[coluna_nos, coluna_velocidade,erro]
                lista_de_lista.append(linha)
                
    df = pd.DataFrame(lista_de_lista, columns=['Qtde_nos', 'Velocidade', 'Erro'])
    
    return df
```

```python
import scipy.stats as st
import numpy as np

def intervalo_confianca(vetor):
    intervalo_confianca = st.t.interval(alpha=0.95, df=len(vetor)-1, loc=np.mean(vetor), scale=st.sem(vetor))
    return (intervalo_confianca[1] - intervalo_confianca[0])
    
    
```
